```python
import pandas as pd

# Load the dataset
file_path = "/content/Business_Management_Dataset (1)_final_v4 (1).csv"
df = pd.read_csv(file_path)

# Display basic information and first few rows
df.info(), df.head()
```

```python
# Filter data for Operations department and Laissez-faire leadership
df_filtered = df[(df["Department"] == "Operations") & (df["Leadership_Style"] == "Laissez-faire")]

# Generate descriptive statistics for Employee Satisfaction and Employee Turnover
desc_stats = df_filtered[["Employee_Satisfaction", "Employee_Turnover"]].describe()

# Compute correlation between Employee Satisfaction and Employee Turnover
correlation = df_filtered["Employee_Satisfaction"].corr(df_filtered["Employee_Turnover"])

desc_stats, correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for Operations department in Asia under Laissez-faire leadership
df_asia = df_filtered[df_filtered["Region"] == "Asia"]

# Scatter plot with regression line
plt.figure(figsize=(8, 5))
sns.regplot(x=df_asia["Employee_Satisfaction"], y=df_asia["Employee_Turnover"], scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})
plt.title("Employee Satisfaction vs Turnover Rate (Laissez-faire Leadership, Operations, Asia)")
plt.xlabel("Employee Satisfaction")
plt.ylabel("Employee Turnover Rate (%)")
plt.grid(True)
plt.show()
```

```python
# Filter data for Finance department in North America under Transformational leadership
df_finance_na = df[(df["Department"] == "Finance") &
                    (df["Region"] == "North America") &
                    (df["Leadership_Style"] == "Transformational")]

# Generate descriptive statistics for Training Hours and Productivity
desc_stats_finance = df_finance_na[["Training_Hours", "Productivity"]].describe()

# Compute correlation between Training Hours and Productivity
correlation_finance = df_finance_na["Training_Hours"].corr(df_finance_na["Productivity"])

desc_stats_finance, correlation_finance
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the training hours and productivity relationship under transformational leadership
# Training hours (X-axis)
training_hours = np.array([10, 20, 30, 40, 50, 60, 70])

# Productivity (Y-axis) - This is an example, assuming productivity increases with training and is influenced by leadership
productivity_transformational_leadership = np.array([40, 50, 65, 75, 85, 90, 95])

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(training_hours, productivity_transformational_leadership, color='b', marker='o', label="Transformational Leadership", linestyle='-', linewidth=2)

# Add titles and labels
plt.title('Relationship Between Training Hours and Productivity with Transformational Leadership', fontsize=14)
plt.xlabel('Training Hours', fontsize=12)
plt.ylabel('Productivity', fontsize=12)

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/Business_Management_Dataset (1)_final_v4 (1).csv'
df = pd.read_csv(file_path)

# Filter the dataset for Authoritarian leadership in the Finance department
authoritarian_finance = df[(df['Leadership_Style'] == 'Autocratic') & (df['Department'] == 'Finance')]

# Calculate the average turnover rate by region
turnover_by_region = authoritarian_finance.groupby('Region')['Employee_Turnover'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(turnover_by_region.index, turnover_by_region.values, color='darkblue')

# Add labels and title
plt.title('Average Turnover Rate in Finance Department under Authoritarian Leadership by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Turnover Rate', fontsize=12)

# Display the value on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, round(yval, 2), ha='center', va='bottom', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
